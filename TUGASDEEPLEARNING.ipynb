```python
# =========================================================
# KLASIFIKASI MOBIL vs MOTOR (FINAL PROJECT)
# =========================================================

# =========================================================
# 1. IMPORT LIBRARY
# =========================================================
import os
import cv2
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import pandas as pd
from glob import glob
import requests
from bs4 import BeautifulSoup

np.random.seed(42)
random.seed(42)

IMG_SIZE = 64
LR = 0.001

# =========================================================
# LOSS FUNCTION
# =========================================================
def binary_cross_entropy(y_true, y_pred):
    y_pred = np.clip(y_pred, 1e-7, 1-1e-7)
    return -(y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))

# =========================================================
# 2. SCRAPING DATASET
# =========================================================
def download_images(keyword, save_folder, max_images=300):

    os.makedirs(save_folder, exist_ok=True)

    url = f"https://www.bing.com/images/search?q={keyword}"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    imgs = soup.find_all("img")

    count = 0

    for img in imgs:
        try:
            src = img.get("data-src") or img.get("src")

            if src is None or not src.startswith("http"):
                continue

            img_data = requests.get(src).content

            with open(f"{save_folder}/{count}.jpg", "wb") as f:
                f.write(img_data)

            count += 1
            if count >= max_images:
                break

        except:
            continue

    print(f"{keyword} downloaded:", count)

# ambil dataset
download_images("car side view", "dataset/Car")
download_images("motorbike side view", "dataset/Bike")

# =========================================================
# DATA AUGMENTATION
# =========================================================
def augment_image(img):

    aug = []
    aug.append(img)
    aug.append(cv2.flip(img, 1))
    aug.append(cv2.flip(img, 0))

    M1 = cv2.getRotationMatrix2D((IMG_SIZE//2, IMG_SIZE//2), 15, 1)
    aug.append(cv2.warpAffine(img, M1, (IMG_SIZE, IMG_SIZE)))

    M2 = cv2.getRotationMatrix2D((IMG_SIZE//2, IMG_SIZE//2), -15, 1)
    aug.append(cv2.warpAffine(img, M2, (IMG_SIZE, IMG_SIZE)))

    bright = np.clip(img + np.random.uniform(-0.1, 0.1), 0, 1)
    aug.append(bright)

    return aug

# =========================================================
# VISUALISASI AUGMENTASI
# =========================================================
def show_augmentation_sample(img):

    aug = augment_image(img)

    plt.figure(figsize=(10,4))
    titles = ["Original", "Flip H", "Flip V", "Rot +15", "Rot -15", "Brightness"]

    for i in range(len(aug)):
        plt.subplot(1,6,i+1)
        plt.imshow(aug[i], cmap="gray")
        plt.title(titles[i])
        plt.axis("off")

    plt.tight_layout()
    plt.show()

# =========================================================
# LOAD DATA
# =========================================================
def load_data(path):

    X, y = [], []
    classes = ["Car", "Bike"]

    for label, cls in enumerate(classes):

        folder = os.path.join(path, cls)
        images = glob(folder + "/*")

        print(f"\n{cls} total:", len(images))

        shown = False

        for img_path in images:

            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue

            if img.shape[0] < 10 or img.shape[1] < 10:
                continue

            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            img = img / 255.0

            if not shown:
                print(f"Contoh Augmentasi: {cls}")
                show_augmentation_sample(img)
                shown = True

            for a in augment_image(img):
                X.append(a)
                y.append(label)

    print("\nTotal setelah augmentasi:", len(X))
    return np.array(X), np.array(y)

# =========================================================
# SPLIT DATA
# =========================================================
def split_data(X, y):

    idx = np.arange(len(X))
    np.random.shuffle(idx)

    X, y = X[idx], y[idx]

    split = int(len(X) * 0.8)

    return X[:split], X[split:], y[:split], y[split:]

# =========================================================
# CNN COMPONENTS
# =========================================================
class Conv2D:
    def __init__(self, filters, k):
        self.k = k
        self.filters = np.random.randn(filters, k, k) * 0.1

    def forward(self, x):

        if len(x.shape) == 2:
            x = x[np.newaxis, :, :]

        c, h, w = x.shape
        out = np.zeros((len(self.filters), h-self.k+1, w-self.k+1))

        for f in range(len(self.filters)):
            for i in range(h-self.k+1):
                for j in range(w-self.k+1):
                    region = x[0, i:i+self.k, j:j+self.k]
                    out[f,i,j] = np.sum(region * self.filters[f])

        return out

class ReLU:
    def forward(self, x):
        return np.maximum(0, x)

class Sigmoid:
    def forward(self, x):
        return 1/(1+np.exp(-x))

class Tanh:
    def forward(self, x):
        return np.tanh(x)

class MaxPool:
    def forward(self, x):

        f,h,w = x.shape
        out = np.zeros((f,h//2,w//2))

        for i in range(f):
            for j in range(h//2):
                for k in range(w//2):
                    out[i,j,k] = np.max(x[i,j*2:j*2+2,k*2:k*2+2])

        return out

class Flatten:
    def forward(self, x):
        return x.flatten()

class Dense:
    def __init__(self, n):
        self.w = np.random.randn(n,1) * 0.1
        self.b = 0

    def forward(self, x):
        self.x = x.reshape(-1,1)
        self.out = np.dot(self.x.T, self.w) + self.b
        return self.out[0][0]

    def backward(self, grad):
        self.w -= LR * (self.x * grad)
        self.b -= LR * grad

# =========================================================
# CNN MODEL
# =========================================================
class CNN:
    def __init__(self, conv_layers=1, activation="relu"):

        self.conv_layers = [Conv2D(4*(i+1),3) for i in range(conv_layers)]

        if activation == "relu":
            self.act = ReLU()
        elif activation == "sigmoid":
            self.act = Sigmoid()
        else:
            self.act = Tanh()

        self.pool = MaxPool()
        self.flat = Flatten()
        self.fc = None

    def forward(self, x):

        for c in self.conv_layers:
            x = c.forward(x)
            x = self.act.forward(x)
            x = self.pool.forward(x)

        x = self.flat.forward(x)

        if self.fc is None:
            self.fc = Dense(len(x))

        x = self.fc.forward(x)
        self.out = 1/(1+np.exp(-x))

        return self.out

    def backward(self, y):
        grad = self.out - y
        self.fc.backward(grad)

# =========================================================
# TRAIN
# =========================================================
def train(model, X, y, epochs=10):

    for e in range(epochs):
        loss_total = 0

        for i in range(len(X)):
            pred = model.forward(X[i])
            loss = binary_cross_entropy(y[i], pred)
            loss_total += loss
            model.backward(y[i])

        print(f"Epoch {e+1} Loss: {loss_total/len(X):.4f}")

# =========================================================
# EVALUATE
# =========================================================
def evaluate(model, X, y):

    correct = 0

    for i in range(len(X)):
        pred = model.forward(X[i])
        pred = 1 if pred > 0.5 else 0

        if pred == y[i]:
            correct += 1

    acc = correct / len(X)
    print("Accuracy:", acc)
    return acc

# =========================================================
# CONFUSION MATRIX
# =========================================================
def confusion_matrix(model, X, y):

    tp = tn = fp = fn = 0

    for i in range(len(X)):
        pred = model.forward(X[i])
        pred = 1 if pred > 0.5 else 0

        if pred == 1 and y[i] == 1:
            tp += 1
        elif pred == 0 and y[i] == 0:
            tn += 1
        elif pred == 1 and y[i] == 0:
            fp += 1
        elif pred == 0 and y[i] == 1:
            fn += 1

    cm = np.array([[tn, fp],
                   [fn, tp]])

    print("\nConfusion Matrix:")
    print(cm)

    return cm

# =========================================================
# VISUALISASI CONFUSION MATRIX
# =========================================================
def plot_confusion_matrix(cm, title="Confusion Matrix"):

    plt.figure()
    plt.imshow(cm)
    plt.title(title)
    plt.colorbar()

    labels = ["Car", "Bike"]

    plt.xticks([0,1], labels)
    plt.yticks([0,1], labels)

    # isi angka di dalam kotak
    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i, j],
                     ha="center", va="center")

    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

# =========================================================
# ABLATION
# =========================================================
def run_ablation(X_train, y_train, X_test, y_test):

    configs = [
        (1,"relu"), (2,"relu"), (3,"relu"),
        (1,"sigmoid"), (2,"sigmoid"), (3,"sigmoid"),
        (1,"tanh"), (2,"tanh"), (3,"tanh")
    ]

    results = []

    for conv, act in configs:

        print("\n====================")
        print(conv,"Conv +",act)
        print("====================")

        model = CNN(conv, act)

        start = time.time()
        train(model, X_train, y_train)
        end = time.time()

        acc = evaluate(model, X_test, y_test)
        cm = confusion_matrix(model, X_test, y_test)
        plot_confusion_matrix(cm, title=f"{conv} Conv + {act}")

        results.append({
            "conv": conv,
            "act": act,
            "acc": acc,
            "time": end-start
        })

    return results

# =========================================================
# MAIN
# =========================================================
X,y = load_data("dataset")
X_train,X_test,y_train,y_test = split_data(X,y)

print("\nTrain:", len(X_train))
print("Test :", len(X_test))

results = run_ablation(X_train,y_train,X_test,y_test)

# =========================================================
# TABLE + GRAPH + ANALISIS
# =========================================================
df = pd.DataFrame(results)
df["model"] = df["conv"].astype(str) + "-" + df["act"]

print("\n===== TABLE =====")
print(df[["model","acc","time"]])

plt.figure()
plt.bar(df["model"], df["acc"])
plt.title("Accuracy Comparison")
plt.xticks(rotation=45)
plt.show()

plt.figure()
plt.bar(df["model"], df["time"])
plt.title("Training Time Comparison")
plt.xticks(rotation=45)
plt.show()

print("\n===== ANALISIS =====")
best = df.loc[df["acc"].idxmax()]
fastest = df.loc[df["time"].idxmin()]

print("Model terbaik:", best["model"])
print("Model tercepat:", fastest["model"])

# =========================================================
# SIMPAN DATASET
# =========================================================
import shutil
from google.colab import files

shutil.make_archive("dataset", 'zip', "dataset")
files.download("dataset.zip")
```
